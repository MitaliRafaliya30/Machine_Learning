```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
print(iris.DESCR)
```

```python
print(iris.target)
```

```python
print(iris.target_names)
```

```python
x = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
y = iris.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

# decision tree do not need for feature scaling 

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# set max_depth=2 for pre-pruning
model = DecisionTreeClassifier(max_depth=2)
```

```python
model.fit(x_train,y_train)
```

```python
#visulization
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(model, filled=True)
```

```python
y_pred = model.predict(x_test)
```

```python
# check performance 
```

```python
# without max_depth defining
model_depth = DecisionTreeClassifier()
model_depth.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(model_depth, filled=True)
```

# pre-pruning or hyperparameter tuning :

```python
param = { 'criterion' : ['gini', 'entropy', 'log_loss'],
         'splitter' : ['best', 'random'],
         'max_depth' : [1,2,3,4,5],
         'max_features' : ['auto', 'sqrt', 'log2'],
}
```

```python
param
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(model_depth, param, cv=5, scoring='accuracy')
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```
